# Task 2 - Huffman coding function

This is the exciting part. Fill in the Huffman coding function `huffman_codebook(freq_dict)`. It takes in the frequency distribution you just created and returns another dictionary showing the codebook. Just a reminder, a codebook just shows the mapping of the source symbols to the Huffman codewords.

THIS TABLE IS JUST AN EXAMPLE. IT'S NOT THE CORRECT ONE.

Symbol             | Codewords        | 
-------------------|------------------|
a                  |  1011            |
b                  |  10              |
c                  |  10110           |
d                  |  100             |

The output dictionary should have something like:

`{'a':'1011', 'b':'10', 'c':'10110', 'd':'100'}`

Where the binary code are in strings. We'll use these for decoding hidden messages later.

A few guidelines:
* Make sure the Huffman coding follows the discussion from our Wiki page
* All left branches are tagged with 0s while all right branches are tagged with 1s
* In case of equal probabilities, the deeper tree takes the left path. If they have the same depth, then sort alphabetically
* Actually... Python dictionaries already sort keys first by string length, then by alphabetical order. So you wouldn't worry too much



```python
###############################################
# Huffman coding function
###############################################
def sort_pair(x):
  x_sorted = dict(sorted(x.items(), key=lambda item: item[1])) #sort dict by frequencies in ascending order
  tuples_list = [(key,value) for key, value in x_sorted.items()] #convert to list of tuples
  #print(tuples_list)
  return tuples_list[0][0], tuples_list[1][0] #return first two items which have the lowest values of freqs 

def huffman_codebook(freq_dict):
  if(len(freq_dict) == 2):
      return dict(zip(freq_dict.keys(), ['0', '1']))

  x_copy = freq_dict.copy()
  pair1,pair2 = sort_pair(freq_dict) #keys of chars of lowest probabilities
  x_copy[pair1 + pair2] = x_copy.pop(pair1) + x_copy.pop(pair2) #combined pair with two lowest probabilities and add entry to dic   

 # print(x_copy)
  codebook = huffman_codebook(x_copy)
  branch = codebook.pop(pair1 + pair2)
  #print(pair1, pair2)
  codebook[pair1], codebook[pair2] = branch + '0', branch + '1' #assign 0 to left branch, assign 1 to right branch
  #print(codebook)
  #print(codebook[pair1],codebook[pair2])
  
  
  return codebook

print(huffman_codebook(freq_dict))
```

In this task you'll need to create a function `decode_message(decodebook,text)`.

*The text argument is a string of binary symbols (e.g.,`1110101001`). The function does not return anything except that it prints the decoded message.
